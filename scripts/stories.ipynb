```python
import json

# Define the Story class
class Story:
    def __init__(self, text, emotions, correct_emotion, image_path):
        self.text = text
        self.emotions = emotions
        self.correct_emotion = correct_emotion
        self.image_path = image_path

# Function to load the JSON file and create Story objects
def load_stories_from_json(filename="stories_data.json"):
    with open(filename, 'r') as file:
        data = json.load(file)  # Load the JSON data into a dictionary

    # Create Story objects from the loaded data
    stories = []
    for item in data:
        story = Story(
            text=item['story'],
            emotions=item['options'],
            correct_emotion=item['correct_emotion'],
            image_path="images/placeholder.png"  # You can adjust the image path as needed
        )
        stories.append(story)
    
    return stories

# Function to extract stories that have emotions in emotion_colors
def extract_stories_in_emotions(stories, target_emotions):
    filtered_stories = []

    # Iterate over each story and check if it contains any of the target emotions
    for story in stories:
        if any(emotion in target_emotions for emotion in story.emotions):
            filtered_stories.append({
                'story': story.text,
                'options': story.emotions,
                'correct_emotion': story.correct_emotion
            })

    return filtered_stories

# List of emotions to filter stories by (emotions in emotion_colors)
emotion_colors = {
    "Joyful": "#e0c158",      # Yellow
    "Sad": "#6da7d1",         # Blue
    "Angry": "#ce8271",       # Redish/Pink
    "Scared": "#59b59d",        # Turquoise/Darkish green
    "Confident": "#d3e7b3",   # Light green
    "Happy": "#edf5e0",       # Very Light Green
    "Frustrated": "#daa194",  # Lighter redish/pink
    "Tired": "#cad4e1",       # Paleish Blue
    "Neutral": "#ffffff"      # White
}

# Load stories from JSON file
stories = load_stories_from_json()

# Extract stories that have any of the emotions in emotion_colors
target_emotions = emotion_colors.keys()  # All emotions in emotion_colors
filtered_stories = extract_stories_in_emotions(stories, target_emotions)

# Save the filtered stories to a new JSON file
with open("filtered_stories_in_emotions.json", "w") as outfile:
    json.dump(filtered_stories, outfile, indent=4)

print("Filtered stories (in emotion_colors) saved to 'filtered_stories_in_emotions.json'.")
```
